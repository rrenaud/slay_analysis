```python
import os, gzip, json, pprint
```

```python
partial_path = 'Monthly_2020_11'
file_list = os.listdir(partial_path)
print(file_list[0])
```

```python
contents = json.load(gzip.open(partial_path + '/' + file_list[0]))
```

```python
for game_log in contents:
    if 'Empty Cage' in game_log['event']['relics']:
        break
else:
    raise ValueError("no such log found")
    
pprint.pprint(game_log)
```
